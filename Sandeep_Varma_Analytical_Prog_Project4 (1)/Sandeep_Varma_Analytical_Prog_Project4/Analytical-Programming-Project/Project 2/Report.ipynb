## CountryScraper Class

The `CountryScraper` class is designed to extract information about countries from a given webpage. It collects data such as country names, capitals, populations, and areas. The output is a pandas DataFrame that can easily be analyzed or saved to a CSV file.

## Extended Data Analysis

Further analysis of the dataset will provide more insights into the performance of the teams. We will look into goal differentials, compare performances to league averages, and visualize the data to uncover trends and patterns.


```python
# Convert 'Goals For (GF)' and 'Goals Against (GA)' columns to numeric types
table_data['Goals For (GF)'] = pd.to_numeric(table_data['Goals For (GF)'])
table_data['Goals Against (GA)'] = pd.to_numeric(table_data['Goals Against (GA)'])

# Calculate goal differential and add it as a new column
table_data['Goal Differential'] = table_data['Goals For (GF)'] - table_data['Goals Against (GA)']

# Display the teams with the highest goal differentials in 1990
top_goal_diff_1990 = table_data[table_data['Year'] == '1990'].sort_values(by='Goal Differential', ascending=False)
top_goal_diff_1990.head()
```

```python
# Calculate the league averages for 1990
average_win_percentage_1990 = table_data[table_data['Year'] == '1990']['Win %'].mean()
average_goal_diff_1990 = table_data[table_data['Year'] == '1990']['Goal Differential'].mean()

print(f"Average win percentage in 1990: {average_win_percentage_1990:.2f}")
print(f"Average goal differential in 1990: {average_goal_diff_1990:.2f}")

# Compare each team's performance to the league average
table_data['Above Average Win %'] = table_data['Win %'] > average_win_percentage_1990
table_data['Above Average Goal Diff'] = table_data['Goal Differential'] > average_goal_diff_1990

# Display the comparison
table_data[table_data['Year'] == '1990'][['Team Name', 'Above Average Win %', 'Above Average Goal Diff']].head()
```
